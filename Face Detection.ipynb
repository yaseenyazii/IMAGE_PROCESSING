So what is Haar Cascade? It is an Object Detection Algorithm used to identify faces in an image or a real time video. 

<b>Note: This is face detection NOT face recognition. We are only detecting if a face is in an image, not who the face actually is.</b>

### Import necessary libraries

## Cascade files

OpenCV comes with these pre-trained cascade files.

### Face Detection

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    
    face_img = img.copy()
    # You may want to play with the scaleFactor and minNeighbor numbers to get good results.
    face_rects = face_cascade.detectMultiScale(face_img, scaleFactor = 1.2, minNeighbors = 2) 
    
    for (x,y,w,h) in face_rects: 
        cv2.rectangle(face_img, (x,y), (x+w,y+h), (255,255,255), 2) 
        
    return face_img
```

```python
result = detect_face(einstein)
```

```python
plt.imshow(result, cmap = 'gray')
```

```python
# Gets errors!
result = detect_face(teresa)
plt.imshow(result, cmap= 'gray')
```

```python
result = detect_face(solvay)
plt.imshow(result,cmap='gray')
```

### Eye Cascade file

```python
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
```

```python
def detect_eyes(img):
    
    face_img = img.copy()
  
    eyes = eye_cascade.detectMultiScale(face_img, scaleFactor = 1.2, minNeighbors = 2) 
    
    
    for (x,y,w,h) in eyes: 
        cv2.rectangle(face_img, (x,y), (x+w,y+h), (255,255,255), 3) 
        
    return face_img
```

```python
result = detect_eyes(teresa)
plt.imshow(result, cmap = 'gray')
```

```python
# White around the pupils is not distinct enough to detect Einstein's eyes here!
result = detect_eyes(einstein)
plt.imshow(result, cmap = 'gray')
```
